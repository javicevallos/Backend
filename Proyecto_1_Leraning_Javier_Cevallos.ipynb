# Ejercicio N° 1
Escribir un programa que pregunte al usuario o genere las ventas de los últimos 5 días, y muestre por pantalla una serie (Dataframe) con los datos de las ventas por producto (opcional): 

    * Nro de unidades y precio unitario, indexada por los días, antes y después de aplicarles un descuento del 5%.
(Diversas opciones de solución: libre)

## Solución
Para la dar una solución al caso planteado, se procedió en primer lugar a importar las librerías a utilizar (panda y datatime) y se crea un DataFrame vacío. 
Posterior a ello, se crean variables para capturar la información tanto del producto, cantidad vendida, y precio unitario. Vale agregar que la fecha se agrega de manera automática al dataframe.


### Código

```python
# Importando librería de Pandas para trabajar con DataFrame
import pandas as pd
import numpy as np

# librería para el ingreso de la fecha
from datetime import datetime 

#creando DataFrame vacío
df = pd.DataFrame()
```

```python
producto = str(input("Ingrese el nombre del artículo: "))
```

```python
producto = str(input("Ingrese el nombre del artículo: "))
```

```python
cantidad = int(input("Ingrese la cantidad: "))
```

```python
ventas = int(input("Precio unitario: "))

# Creación de variables y asignaciones
hoy=datetime.now()
hoy = hoy.strftime("%d %b %Y %H:%M:%S") #formato de la fecha día/mes/año/hora/minuto/segundos
df['fecha'] = hoy
df['producto'] = producto
df['cantidad'] = cantidad
df['ventas'] = ventas

# incorporando una nueva fila al DataFrame
nueva_fila = { 'fecha': hoy,
              'producto': producto,
              'cantidad': cantidad,
              'ventas': ventas,
              'Antes' : cantidad*ventas,
             'Descontado':cantidad*ventas*0.95} # creamos un diccionario
df = df.append(nueva_fila, ignore_index=True)

#indexando la fecha
df.set_index('fecha', inplace = True)
```

# Ejercicio Nº 3
Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:

```python
mes = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
Ventas = [30500, 35600, 28300, 33900, 42500]
Gastos = [22000, 23450, 18100, 35700, 32450]
tabla = pd.DataFrame(columns=['mes', 'Ventas', 'Gastos'])
tabla['mes'] = mes
tabla['Ventas'] = Ventas
tabla['Gastos'] = Gastos
tabla
```

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

## Solución
Para la 'Columna1' se utiliza una división de DataFrame tabla, campo 'Gastos' con el campo 'Ventas'
### Columna 1

```python
tabla['Columna1']=tabla['Gastos']/tabla['Ventas']*100

tabla
```

### Columna 2

Para la columna 2, se utiliza un for para que vaya buscando (loc) en cada fila de la columna *Ventas* y compara los valores utilizándose una función *if*

```python
#Generando la columna 2
for i in range(0 , len(tabla)):
    if tabla.loc[i,'Ventas'] >= 30000 :    
        tabla.loc[i,'Columna2'] = "Meta Superada"
    else:
        tabla.loc[i,'Columna2'] = "Meta NO Superada"

tabla
```

### Columna 3

Se pide que la Columna3 calcule el sobre gastos siempre y cuando el Gasto se mayor que las Ventas

```python
#Generando la columna 3
for i in range(0 , len(tabla)):
    if tabla.loc[i,'Gastos'] >  tabla.loc[i,'Ventas']:    
        tabla.loc[i,'Columna3'] = tabla.loc[i,'Gastos'] - tabla.loc[i,'Ventas']
    else:
        tabla.loc[i,'Columna3'] = "-"*3

tabla
```

### Obtener el total de las ventas de los últimos 3 meses
La nueva tabla total será el DataFrame que corresponde el largo de la tabla menos 3, hasta el largo de la tabla

```python
total=tabla[len(tabla)-3:len(tabla)] #Últimos 3 meses
total
```

```python
print("La suma de ventas de los últimos 3 meses es: ",total['Ventas'].sum())
```

### Obtener el total de gastos de febrero a abril

```python
total2=tabla[1:4]
total2
```

```python
print("La suma de ventas desde febrero hasta Abril es: ",total2['Ventas'].sum())
```
