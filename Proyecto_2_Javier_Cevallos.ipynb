### Función asociada con Raíz cuadrada y exponencial 
Para el presente ejercicio se utiliza como variables de estudio *'Millas_Por_Galon' y *'Cilindros'* como var1 y var2, respectivamente. Este cambio se debe a que al usar la variable anterior de var2, los números generados son extremadamente grandes, imposibilitando la visualización del cálculo generado.

```python
#Sqrt(var1)*exp(var2)/200

def raizcuadrada(x: float, y: float, TypeCal: bool=False) -> float:
    '''
    Función para la raíz cuadrada de una variable y multiplicarla por e elevado a la segunda variable y 
    dividirla por una constante (200)

    '''
      
    rc = np.sqrt(x)*np.exp(y)/200
    return rc
```

### Función para condicionante entre variables

```python
#Si var 1 > var 2 → 5, caso contrario 3
def cond_variable (x: float, y: float, TypeCal: bool=False) -> int:
    import numpy as np
        
    '''
    Función que permite comparar dos variables bajo la condición de que 
    si var1 es mayor que var2, entonces la función retorna un número 5, 
    pero si esto no se cumple retornará un 3
    fuente: https://statologos.com/comparar-dos-columnas-en-pandas/
    '''
    condiciones = [x > y, x <= y]
    opciones = [5,3]
    resultado = np.select(condiciones, opciones) 
    return resultado  
```

```python
var_3 = cond_variable(x = df['Millas_Por_Galon'],
                      y = df['Aceleracion'],
                      TypeCal = True)
var_3 = pd.Series(var_3) #transformándolo a una serie
var_3.head (35)
```

### Operatoria matemática $\frac{1}{log(var_1/var_2)}*100$

```python
### 1/logn(var1/var2)*100
def log_n (n:int, x: float, y: float, TypeCal: bool=False) -> float:
    """
    Función que permite calcular una división con logaritmos natural dos variables (var1 - var2) 
    """
    return (1/np.log(x/y)*100)
```

```python
#utilizando la función creada con los parámetros definidos
var_4 = log_n(n=5, x = df['Millas_Por_Galon'],
                      y = df['Cilindros'],
                      TypeCal=True)
var_4
```

### Operatoria matemática $(var_2)^2/var_1$

```python
### var2**2/var1
def v_cuadrado (x: float, y: float, TypeCal: bool=False) -> float:
    """
    Función que permite recibir dos variables (var1 - var2) y eleva al cuadrado la segunda
    variable y lo divide por var1
    """
    return y**2/x
```

```python
var_5 = v_cuadrado(x = df['Millas_Por_Galon'],
                      y = df['Cilindros'],
                      TypeCal=True)
var_5
```

# Ejercicio Nº 3
- Crear una función que permita generar 3 variables cualitativas ordinales a partir de las variables numéricas con los siguientes criterios:  

★ Si es mayor a la “media + 1xdesviación estándar” → **Alto**   
★ Si está entre “media - 1xdesviación estándar” y “media + 1xdesviación estándar” → **Medio**  
★ Si es menor a la “media - 1xdesviación estándar”  → **Bajo**  

Nota: Mostrar con las nuevas variables generadas.  
Elegir con libertad las variables a trabajar.  
La función debe contar con su respectiva documentación  

## Solución

```python
def media_sd (x:float, media=float, desv=float, TypeCal:bool = False) -> float:
    """
    Permite categorizar un valor según el promedio y desviación estándar asociándolo a 3 posibles valores
    Bajo, medio o Alto
    ★ Si es mayor a la “media + 1xdesviación estándar” → Alto
    ★ Si está entre “media - 1xdesviación estándar” y “media + 1xdesviación estándar” → Medio
    ★ Si es menor a la “media - 1xdesviación estándar” → Bajo
    
    -------
    Variables
    x = vector 
    media: recibe el promedio del vector
    desv: recibe la desviación estándar del vector
    TypeCal
    """
    condiciones = [(x > (media + desv)), x < (media - desv), ((x <= (media + desv)) & (x >= (media - desv)))]
    opciones = ["Alto","Bajo", "Medio"]
    resultado = np.select(condiciones, opciones)
    return resultado 
```

```python
categorizacion = media_sd(x = df['Millas_Por_Galon'],
                       media = df['Millas_Por_Galon'].mean(),
                        desv = df['Millas_Por_Galon'].std(),
                        TypeCal = True)
categorizacion = pd.Series(categorizacion)
categorizacion
```

```python
df1["Categorizacion"]=categorizacion
df1.head(10)
```
